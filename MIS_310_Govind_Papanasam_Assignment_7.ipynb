# Part C - General Visualization

```python
# Bar Chart - Average age by membership status

dfpct.groupby("status")["age"].mean().plot(kind="bar")

plt.title("Average Age By Membership Status")
plt.xlabel("Status")
plt.xticks(rotation = 0, ha ="center")
plt.ylabel("Age")
plt.show()
```

```python
# Pie Chart - Distribution of Gender
dfpct["gender"].dropna().value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.ylabel("")
plt.title("Gender Distribution")
plt.show()
```

```python
# Histogram - Distribution Of Age
dfpct["age"].plot(kind="hist", bins = 20)

plt.title("Member Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
